```python
import image_processing as pi
```

```python
# Lendo arquivo
filename = 'images/moon.png'
```

```python
# Lendo imagem com necessidade de processamento
img_matrix = pi.transformImage(filename)
```

```python
# Lendo imagem sem nec. de proc.
img_matrix = pi.readImage(filename)
```

```python
# Exibindo imagem
pi.showImage(img_matrix)
```

```python
# Usando filtro negativo
img_negative = pi.filterNegative(img_matrix)
pi.showImage(img_negative)
```

```python
# Usando filtro de contraste com log
img_contrast_log = pi.filterContrastLog(img_matrix)
pi.showImage(img_contrast_log)
```
